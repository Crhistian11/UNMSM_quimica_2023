```python
#Importar la biblioteca Pandas
import pandas as pd

#Crear los datos
elementos= ["Cl2","N2","He"]
temperaturas = [100,300,700]
velocidades_moleculares = [750,1450,1650]

#Crear el dataframe
df= pd.DataFrame({
    "Elementos": elementos,
    "Temperatura": temperaturas,
    "Velocidad Molecular": velocidades_moleculares
})
#Obtener las medidas de tendencia central
medidas_tendencia_central = df.describe()

#Imprimir las medidas de tendencia central
print(medidas_tendencia_central)
```

```python
import math
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.style as style
import pandas as pd
```

```python
def f(x):
   return 1000 + (1 - (x - 250)**2/(1000**2 * 2))**1/2


x = np.linspace(0, 500, 1000)
y = f(x)
```

```python
plt.plot(x, y)
plt.xlabel("Velocidad molecular(m/s)")
plt.ylabel("Numero de Moleculas")
plt.title("100K")
```
